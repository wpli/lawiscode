```python
years = [ 1925, 1934, 1940, 1946, 1952, 1958, 1964, 1970, 1976, 1982, 1988, 1994, 2000, 2006, 2012 ]
titles = range( 1, 52 )
```

```python
data_tuples = []

year_title_length_dict = {}
with open( 'data/lengths_from_hein.txt' ) as f:
    x = f.read().split('\n')
    if x[-1] == '':
        del x[-1]
    
    for i in x:
        entry = i.split()
        year = int( entry[0] )
        title = int( entry[1] )
        length = int( entry[2] )
        data_tuples.append( ( year, title, length ) )
        
raw_data_matrix = numpy.zeros( ( len( titles ), len( years ) ), dtype=numpy.float )

year_to_idx_dict = dict( [ (year,idx) for (idx,year) in enumerate( years ) ] )
title_to_idx_dict = dict( [ (title,idx) for (idx,title) in enumerate( titles ) ] )

for tup in data_tuples:
    y_idx = year_to_idx_dict[tup[0]]
    t_idx = title_to_idx_dict[tup[1]]
    raw_data_matrix[t_idx,y_idx] = tup[2]
                    

normalized_data_matrix = numpy.zeros( ( len( titles ), len( years ) ), dtype=numpy.float )
for y_idx, year in enumerate( years ):
    for t_idx, title in enumerate( titles ):
        normalized_data_matrix[t_idx, y_idx] = raw_data_matrix[t_idx, y_idx] / (raw_data_matrix[t_idx, 0 ]+1)

```

```python
xhtml_data_tuples = []
with open( 'data/lengths_from_xhtml.txt' ) as f:
    x = f.read().split('\n')
    if x[-1] == '':
        del x[-1]
    
    for i in x:
        entry = i.split()
        year = int( entry[0] )
        title = int( entry[1] )
        length = int( entry[2] )
        xhtml_data_tuples.append( ( year, title, length ) )
```
